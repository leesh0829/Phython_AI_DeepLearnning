```python
# 데이터 로드

import pandas as pd

data = pd.read_csv("Student_Marks.csv")
data
```

### 데이터 탐색 및 전처리

```python
# 데이터 기초 정보 확인하기
# info( ) 메소드를 통해 데이터 개수, 속성 개수, 속성명, 결측치, 속성의 데이터 타입 등 확인

data.info()
```

```python
# 데이터 통계치 살펴보기
# describe( ) 메소드를 통해 데이터 개수, 평균, 표준편차, 최솟값, 최댓값, 4분위수 파악

data.describe()
```

```python
# 상관관계(correlation) 분석

corrlation = data.corr()
corrlation
```

```python
# 상관관계 분석 시각화
# 결과를 히트맵으로 그리기

import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(corrlation,annot=True,cmap='spring')
plt.show()
```

```python
# 산점도로 표현
# 여러 개 그래프 그리기

fig, ax = plt.subplots(ncols=2,figsize=(12,6))
sns.scatterplot(data=data,x = 'number_courses',y='Marks',ax=ax[0])
sns.scatterplot(data=data,x = 'time_study',y='Marks',ax=ax[1])
plt.show()
```

```python
# 훈련데이터와 테스트 데이터 나누기

from sklearn.model_selection import train_test_split

X = data.drop(columns=['Marks'])
Y = data['Marks']


train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size=0.2,random_state=42)

print(train_x.shape,test_x.shape,train_y.shape,test_y.shape)
```

### 모델 생성 및 학습

```python
# 모델 생성

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# 모델 학습
model.fit(train_x,train_y)
```

```python
# 모델 성능 평가 라이브러리 불러오기
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
# 모델 성능 평가

pred = model.predict(test_x)

print(f'r_score: {r2_score(test_y,pred):.2f}')
print(f'Mean Squared Error: {mean_squared_error(test_y,pred):.2f}')
```
