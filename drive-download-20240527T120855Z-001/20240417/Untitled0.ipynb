```python
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow
```

```python
cap = cv2.VideoCapture('01.mp4')

fourcc = cv2.VideoWriter_fourcc('m', 'p', '4', 'v')

#cv2.VideoWrite(outputFile, fourcc, frame, size)

out = cv2.VideoWriter('output.mp4', fourcc, cap.get(cv2.CAP_PROP_FPS),
                      #(int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int (cap.get(cv2.CAP_PROP_FRAME_HEIGHT))))
                      (700, 200))

while cap.isOpened():
  ret, img = cap.read()

  if not ret:
    break

  cv2.rectangle(img, pt1=(200, 200), pt2=(300, 300), color=(255, 255, 255), thickness=2)
  #img = cv2.resize(img, dsize=(640, 360))
  img = img[100:300, 100:800]

  #cv2_imshow(img)
  out.write(img)

files.download('output.mp4')
```
